Now we check for Stationarity 

```python
from statsmodels.tsa.stattools import adfuller

def adf_test(timeseries):
    """
    Conducts an Augmented Dickey-Fuller (ADF) test on a time series
    """
    result = adfuller(timeseries)
    print('ADF Statistic: {:.4f}'.format(result[0]))
    print('p-value: {:.4f}'.format(result[1]))
    print('Critical Values:')
    for key, value in result[4].items():
        print('\t{}: {:.4f}'.format(key, value))
```

```python
adf_test(train)
```

After performing ADF test we got p-value as 0.6085 which is greater than alpha so we fail to reject null hypotheses Therefore we fail to reject Null hypotheses. data is non stationary,we apply 1st order differenciation to make the data from non stationary to staionary.

```python
train_diff=train.diff().dropna()
train_diff
```

Again performimg ADF test to confirm whether the data became staionary or not

Null hypothesis - data is not stationary Alternate hypothesis -data is stationary the data is staionary

```python
adf_test(train_diff)
```

Now we can see that pvalue<alpha.Therefore we fail to accept null hypothese

```python
fig = plt.figure(figsize=(12,8))
ax5= fig.add_subplot(311)
fig=plt.plot(train_diff)

fig = plt.figure(figsize=(12,12))
ax6= fig.add_subplot(312)
fig=sm.graphics.tsa.plot_acf(train_diff,lags=30, ax=ax6)

ax7 = fig.add_subplot(313)
fig = sm.graphics.tsa.plot_pacf(train_diff, lags=30, ax=ax7)
```

```python
# Fit ARIMA model
model143 = sm.tsa.ARIMA(train, order=(0, 1, 0))
result143 = model143.fit()

# Get residuals
#residuals = result.resid
```

```python
# Print model summary
print(result143.summary())

# Make predictions
preds = result143.predict(start='2018-05-25', end='2019-12-31', dynamic=False)

# Plot actual and predicted values
plt.plot(train.index, train.values, label='Actual')
plt.plot(test.index, test.values, label='Actual')               
plt.plot(preds.index, preds.values, label='Predicted')  
plt.legend()
plt.show()   
```

```python
from sklearn.metrics import mean_absolute_error, mean_squared_error
import numpy as np

# calculate MAPE
mape = np.mean(np.abs((np.array(np.exp(test.values)) - np.array(np.exp(preds.values))) / np.array(np.exp(test.values))) * 100)

# calculate MAE
mae = mean_absolute_error(np.exp(test.values), np.exp(preds.values))

# calculate RMSE
rmse = np.sqrt(mean_squared_error(np.exp(test.values), np.exp(preds.values)))

# print the results
print("MAPE:", mape)
print("MAE:", mae)
print("RMSE:", rmse)
```
